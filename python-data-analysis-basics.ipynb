Once we've analyzed the basic characteristics of the input data, we can pre-process it for machine learning algorithms. Typical pre-processing steps include: 
* load the raw dataset
* split the dataset into input and output variables
* transform the input variables into machine-learning format 

Many machine learning algorithms need input data to use the same scale. The scikit-learn Python library has a handy class for this called MinMaxScaler.

```python
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler

array = data.values

# separate array into input and output attributes and rescale the data
X = array[:,0:8]
Y = array[:,8]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)

# summarize the rescaled data
set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

Sometimes we'll need to transform input data to a standard Gaussian distribution with mean of 0 and standard deviation of 1. This is common for regression and linear discriminate analysis. We can use scikit-learn's StandardScaler class.

```python
from sklearn.preprocessing import StandardScaler

# separate array into input and output attributes and rescale the data
X = array[:,0:8]
Y = array[:,8]
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)

# summarize the rescaled data
set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

Another common use case is data normalization. This means rescaling each row to be length 1. This is useful for any algorithms that use weights (e.g., neural networks) and algorithms that use distance (e.g., k-nearest neighbors). We can use scikit-learn's Normalizer class.

```python
from sklearn.preprocessing import Normalizer

# separate array into input and output attributes and rescale the data
X = array[:,0:8]
Y = array[:,8]
scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)

# summarize the rescaled data
set_printoptions(precision=3)
print(normalizedX[0:5,:])
```

Sometimes we need to transform data into binary format, 0's and 1's. We can use scikit-learn's Binarizer class.

```python
from sklearn.preprocessing import Binarizer

# separate array into input and output attributes and rescale the data
X = array[:,0:8]
Y = array[:,8]
binarizer = Binarizer(threshold=0.0).fit(X)
binaryX = binarizer.transform(X)

# summarize the rescaled data
set_printoptions(precision=3)
print(binaryX[0:5,:])
```

Before running our data through machine learning algorithms, we need to select the best features to use in the input data. This will reduce overfitting and training time and improve accuracy.

One common technique for feature selection is univariate selection. This is a method to select those features with the strongest relationship to the output variable. We can use scikit-learn's SelectKBest class.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# extract features
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)

# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)

# summarize selected features
print(features[0:5,:])
```

Another feature selection approach is Recursive Feature Elimination (RFE). Scikit-learn has a class called RFE for this.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# extract features
model = LogisticRegression()
rfe = RFE(model, 3)
fit = rfe.fit(X, Y)

# summarize selected features
print('Num Features: %s' % (fit.n_features_))
print('Selected Features: %s' % (fit.support_))
print('Feature Ranking: %s' % (fit.ranking_))
```

We can also estimate the importance of features with decision trees. Scikit-learn has a class called ExtraTreesClassifier for this.

```python
from sklearn.ensemble import ExtraTreesClassifier
set_printoptions(precision=8)

# extract features
model = ExtraTreesClassifier()
model.fit(X, Y)

# summarize selected features
print(model.feature_importances_)
```
